```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/fifa_data.csv')
df
```

```python
most_players = df['Nationality'].value_counts()
most_players
```

```python
most_players.head().plot(kind = 'bar',color = 'green')
plt.ylabel('number of players')
plt.title('Countries and number of players')
plt.show()
```

```python
df['Wage'] = df['Wage'].replace('€', '').replace('K', '000')
df['Wage'] = df['Wage'].astype(float)
highest_salary_player = df.sort_values(by='Wage', ascending=False)
highest_salary_player[['Name','Wage']].head()
```

```python
df['Wage'].plot(kind='hist',color='green',edgecolor='black')
plt.ylabel('Number of players')
plt.xlabel('Wage(in thousand of euro)')
plt.title('Players and their wages')
plt.show()
```

```python
tallest_player = df.sort_values(by='Height',ascending=False)
tallest_player[['Name','Height']].head()
```

```python
most_players_club = df['Club'].value_counts()
most_players_club.idxmax()
```

```python
preferred_foot = df['Preferred Foot'].value_counts()
preferred_foot
```

```python
preferred_foot.plot(kind='bar',color=('green','red'))
plt.xlabel('preferred foot')
plt.ylabel('Number of players')
plt.title('players & preferred foot')
plt.show()
```
